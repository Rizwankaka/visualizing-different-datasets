## Visualization of tips dataframe using `Pandas`

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load the dataset
df= sns.load_dataset('tips')
df.to_csv('tips.csv')
```

```python
# Plot 1: Histogram of total bill
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
df['total_bill'].hist(edgecolor='black', bins=20)
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

# Plot 2: Histogram of tips
plt.subplot(1, 2, 2)
df['tip'].hist(edgecolor='black', bins=20, color='orange')
plt.title('Histogram of Tip Amount')
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Interpretations:**

- **Histogram of Total Bill:** Most of the total bill amounts are concentrated between 10 and 20 dollars.
- **Histogram of Tip Amount:** The tip amounts are mostly concentrated between 1 and 4 dollars.

```python
# Plot 3: Boxplot of total bills by day
plt.figure(figsize=(15, 5))
df.boxplot(column='total_bill', by='day')
plt.title('Boxplot of Total Bills by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.suptitle('')

# Plot 4: Boxplot of tips by day
df.boxplot(column='tip', by='day', patch_artist=True, boxprops=dict(facecolor="orange"))
plt.title('Boxplot of Tips by Day')
plt.xlabel('Day')
plt.ylabel('Tip Amount')
plt.suptitle('')  

plt.tight_layout()
plt.show()
```

**Interpretations:**

- **Boxplot of Total Bills by Day:** The median of total bill is highest on weekends. There are some outliers, especially on Saturday, indicating that there were a few unusually high bills on that day.
- **Boxplot of Tips by Day:** The median tip amount is also highest on Sundays and Saturdays. Similar to the total bill, there are outliers on Saturday, showing that some unusually high tips were given on that day.

```python
# Plot 5: Scatter plot of total bill vs. tip
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.scatter(df['total_bill'], df['tip'], alpha=0.6, edgecolor='black')
plt.title('Scatter plot of Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# Plot 6: Scatter plot of total bill vs. tip colored by time
plt.subplot(1, 2, 2)
colors = df['time'].map({'Lunch': 'blue', 'Dinner': 'orange'})
plt.scatter(df['total_bill'], df['tip'], c=colors, alpha=0.6, edgecolor='black')
plt.title('Total Bill vs. Tip by Time of Day')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.tight_layout()
plt.show()
```

 **Interpretations:**
- There's a positive correlation between the total bill and the tip amount, which is expected.
- Most of the higher bills and tips are associated with dinner time, as indicated by the orange points.

```python
# Plot 7: Count of meals by day
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
df['day'].value_counts().plot(kind='bar', color='green')
plt.title('Count of Meals by Day')
plt.xlabel('Day')
plt.ylabel('Count')

# Plot 8: Average tip by gender
plt.subplot(1, 2, 2)
df.groupby('sex')['tip'].mean().plot(kind='bar', color='purple')
plt.title('Average Tip by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Tip')

plt.tight_layout()
plt.show()
```

**Interpretation:**

**Count of Meals by Day:** The dataset contains more meals for Saturday followed by Sunday. These appear to be the busiest days for the restaurant.\
**Average Tip by Gender:** Males, on average, tend to give a slightly higher tip compared to females in this dataset.

```python
# Plot 9: Average total bill by time
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
df.groupby('time')['total_bill'].mean().plot(kind='barh', color='cyan')
plt.title('Average Total Bill by Time of Day')
plt.xlabel('Average Total Bill')
plt.ylabel('Time of Day')

# Plot 10: Average tip percentage by smoker status
df['tip_percentage'] = (df['tip'] / df['total_bill']) * 100
plt.subplot(1, 2, 2)
df.groupby('smoker')['tip_percentage'].mean().plot(kind='barh', color='red')
plt.title('Average Tip Percentage by Smoker Status')
plt.xlabel('Average Tip Percentage')
plt.ylabel('Smoker')

plt.tight_layout()
plt.show()
```

**Interpretations:**
**Average Total Bill by Time of Day:** The average total bill is notably higher during dinner compared to lunch. This might be due to more items being ordered during dinner or possibly more expensive menu items.\
**Average Tip Percentage by Smoker Status**: The tip percentage is nearly the same for both non-smokers and smokers. However, non-smokers give a slightly higher average tip percentage than smokers.


## Visualization of iris dataframe using `matplotlib`

```python
# load the dataset 
df = sns.load_dataset('iris')
df.to_csv('iris.csv')
```

```python
# Plot 1: Histogram of sepal length
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.hist(df['sepal_length'], bins=20, color='blue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

# Plot 2: Histogram of sepal width
plt.subplot(1, 2, 2)
plt.hist(df['sepal_width'], bins=20, color='green', edgecolor='black')
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Interpretation:**

**Histogram of Sepal Length:** Most of the iris flowers have a sepal length concentrated between 5 and 7 cm.\
**Histogram of Sepal Width:** The sepal width for the majority of the iris flowers is between 2.5 and 3.5 cm, with a peak around 3 cm.

```python
# Plot 3: Histogram of petal length
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.hist(df['petal_length'], bins=20, color='red', edgecolor='black')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')

# Plot 4: Histogram of petal width
plt.subplot(1, 2, 2)
plt.hist(df['petal_width'], bins=20, color='purple', edgecolor='black')
plt.title('Histogram of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Interpretations:**
**Histogram of Petal Length:** The petal length shows a bimodal distribution, indicating two distinct groups within the data. There are peaks around 1.5 cm (likely corresponding to the setosa species) and 5 cm.\
**Histogram of Petal Width:** The petal width also displays a bimodal distribution, with concentrations around 0.2 cm (again, likely setosa) and 1.5 cm

```python
# Plot 5: Scatter plot of sepal length vs. sepal width
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
colors = df['species'].map({'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'})
plt.scatter(df['sepal_length'], df['sepal_width'], c=colors, edgecolor='black', alpha=0.6)
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Plot 6: Scatter plot of petal length vs. petal width
plt.subplot(1, 2, 2)
plt.scatter(df['petal_length'], df['petal_width'], c=colors, edgecolor='black', alpha=0.6)
plt.title('Scatter Plot of Petal Length vs. Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

plt.tight_layout()
plt.show()
```

**Scatter Plot of Sepal Length vs. Sepal Width:** The setosa species (represented by red points) generally have smaller sepal lengths but wider sepal widths compared to the other species. This distinction allows for easy identification of the setosa species based on sepal dimensions.\
**Scatter Plot of Petal Length vs. Petal Width:** The scatter plot shows clear clusters for each species based on petal dimensions. This indicates that petal length and width are good features to distinguish between the three species of iris flower

```python
# Plot 7: Sepal length distribution by species
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
for species, color in zip(['setosa', 'versicolor', 'virginica'], ['red', 'green', 'blue']):
    subset = df[df['species'] == species]
    plt.hist(subset['sepal_length'], bins=20, color=color, edgecolor='black', alpha=0.6, label=species)
plt.title('Sepal Length Distribution by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.legend()

# Plot 8: Sepal width distribution by species
plt.subplot(1, 2, 2)
for species, color in zip(['setosa', 'versicolor', 'virginica'], ['red', 'green', 'blue']):
    subset = df[df['species'] == species]
    plt.hist(subset['sepal_width'], bins=20, color=color, edgecolor='black', alpha=0.6, label=species)
plt.title('Sepal Width Distribution by Species')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

**Sepal Length Distribution by Species:** The setosa species generally has the smallest sepal lengths, followed by versicolor. The virginica species typically has the largest sepal lengths.\
**Sepal Width Distribution by Species:** While the setosa species tends to have the widest sepals, there's a notable overlap in sepal width between the three species. This suggests that sepal width alone might not be as distinctive as sepal length in differentiating between the species.

```python
# Plot 9: Petal length distribution by species
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
for species, color in zip(['setosa', 'versicolor', 'virginica'], ['red', 'green', 'blue']):
    subset = df[df['species'] == species]
    plt.hist(subset['petal_length'], bins=20, color=color, edgecolor='black', alpha=0.6, label=species)
plt.title('Petal Length Distribution by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.legend()

# Plot 10: Petal width distribution by species
plt.subplot(1, 2, 2)
for species, color in zip(['setosa', 'versicolor', 'virginica'], ['red', 'green', 'blue']):
    subset = df[df['species'] == species]
    plt.hist(subset['petal_width'], bins=20, color=color, edgecolor='black', alpha=0.6, label=species)
plt.title('Petal Width Distribution by Species')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

**Petal Length Distribution by Species:** The petal lengths are distinct for each species. Setosa has the smallest petal lengths, versicolor has medium lengths, and virginica possesses the longest petals. This distinction in petal length can be a strong feature for species identification.\
**Petal Width Distribution by Species:** Similar to petal length, petal width also provides clear distinctions between the species. Setosa flowers typically have the narrowest petals, while virginica flowers have the widest.

## Visualization of diamonds dataframe using `seaborn`

```python
# load the dataset
df = sns.load_dataset('diamonds')
df.to_csv('diamonds.csv')
```

```python
# Plot 1: Distribution of diamond prices
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['price'], kde=True, color='blue', bins=30)
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Plot 2: Boxplot of price by cut of the diamond
plt.subplot(1, 2, 2)
sns.boxplot(x='cut', y='price', data=df, palette='viridis')
plt.title('Price Distribution by Diamond Cut')
plt.xlabel('Cut')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

**Interpretation:**

**Distribution of Diamond Prices:** The majority of diamonds are priced in the lower range, with a few extending into the higher price brackets, leading to a right-skewed distribution.\
**Price Distribution by Diamond Cut:** Although the median price remains relatively consistent across different cuts, the "Premium" and "Fair" cuts exhibit higher price outliers. This suggests that while cut plays a role in pricing, other factors may also contribute to the higher prices observed for certain diamonds.


```python
# Plot 3: Distribution of diamond carat
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['carat'], kde=True, color='green', bins=30)
plt.title('Distribution of Diamond Carat')
plt.xlabel('Carat')
plt.ylabel('Frequency')

# Plot 4: Boxplot of carat by clarity of the diamond
plt.subplot(1, 2, 2)
sns.boxplot(x='clarity', y='carat', data=df, palette='coolwarm', order=['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1'])
plt.title('Carat Distribution by Diamond Clarity')
plt.xlabel('Clarity')
plt.ylabel('Carat')

plt.tight_layout()
plt.show()
```

**Interpretations:**

**Distribution of Diamond Carat:** The majority of diamonds in the dataset have a size less than 1.5 carats, with a significant peak observed around 0.3 carats. This suggests that smaller diamonds are more commonly available or popular.\
**Carat Distribution by Diamond Clarity:** Diamonds with "I1" clarity (the lowest clarity grade in the dataset) tend to have a higher median carat size. In contrast, diamonds with higher clarity grades, such as "IF" (internally flawless), tend to be smaller. This might indicate a trade-off between size and clarity in the diamond market.

```python
# Plot 5: Count of diamonds by cut
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='cut', data=df, palette='pastel', order=df['cut'].value_counts().index)
plt.title('Count of Diamonds by Cut')
plt.xlabel('Cut')
plt.ylabel('Count')

# Plot 6: Count of diamonds by clarity
plt.subplot(1, 2, 2)
sns.countplot(x='clarity', data=df, palette='husl', order=['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1'])
plt.title('Count of Diamonds by Clarity')
plt.xlabel('Clarity')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

**Interpretation:**

**Count of Diamonds by Cut:** The "Ideal" cut is the most prevalent in the dataset, followed by "Premium" and "Very Good". This suggests that "Ideal" cut diamonds are either more popular or more commonly available in the market.\
**Count of Diamonds by Clarity:** "SI1" and "VS2" are the most common clarity grades in the dataset. These grades indicate slight inclusions that are noticeable to a skilled grader but might not be obvious to the naked eye. On the other hand, "IF" (internally flawless) is the least common clarity grade, which is expected given its rarity.

```python
# Plot 7: Violin plot of price by cut
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.violinplot(x='cut', y='price', data=df, palette='rainbow')
plt.title('Price Distribution by Diamond Cut')
plt.xlabel('Cut')
plt.ylabel('Price')

# Plot 8: Violin plot of price by clarity
plt.subplot(1, 2, 2)
sns.violinplot(x='clarity', y='price', data=df, palette='cool', order=['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1'])
plt.title('Price Distribution by Diamond Clarity')
plt.xlabel('Clarity')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

**Interpretation:**

**Price Distribution by Diamond Cut (Violin Plot):** While there's a significant overlap in price distributions across different diamond cuts, diamonds with the "Fair" and "Good" cuts display a broader range of prices.\
**Price Distribution by Diamond Clarity (Violin Plot):** The price distributions for different clarity grades also overlap. However, diamonds with "IF" (internally flawless) and "I1" (lowest clarity) grades have notably higher price points. This might indicate a premium for flawless diamonds and a price bump for larger diamonds with lower clarity.

```python
# Plot 9: Scatter plot of carat vs. price colored by clarity
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
palette = sns.cubehelix_palette(8, start=.5, rot=-.75)
sns.scatterplot(x='carat', y='price', hue='clarity', data=df, palette=palette, alpha=0.6)
plt.title('Carat vs. Price colored by Clarity')
plt.xlabel('Carat')
plt.ylabel('Price')

# Plot 10: Correlation heatmap of numeric features
plt.subplot(1, 2, 2)
correlation = df.corr()
sns.heatmap(correlation, cmap='coolwarm', annot=True, vmin=-1, vmax=1)
plt.title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```

**Interpretations:**

**Carat vs. Price colored by Clarity:** As the carat size of a diamond increases, its price generally rises. The variation in price for the same carat size is influenced by the clarity of the diamond. For instance, diamonds with lower clarity tend to have lower prices compared to diamonds of the same size but with higher clarity.\
**Correlation Heatmap:** The heatmap shows the correlation between numeric features in the dataset. Carat size has a strong positive correlation with price, indicating that larger diamonds tend to be more expensive. On the other hand, "depth" and "table" have weaker correlations with other features, suggesting that they might not be as influential in determining diamond prices.

## Visualization of car crashes  dataframe using `plotly`

```python
# load the dataset
df= sns.load_dataset('car_crashes')
df.to_csv('car_crashes.csv')
```

```python
import plotly.express as px
# plot 1: Bar chart of total car crashes by state:
fig1 = px.bar(df, x='abbrev', y='total', title='Total Car Crashes by State', color='total', labels={'abbrev': 'State', 'total': 'Total Crashes'})
fig1.show()
```

**Interpretation:**

The bar chart would provide a visual representation of the total number of car crashes by state. States with a darker color would have a higher number of crashes, allowing for an easy comparison between states.

```python
# plot 2: Scatter plot of alcohol-related crashes vs. speeding-related crashes
fig2 = px.scatter(df, x='alcohol', y='speeding', title='Alcohol-related vs. Speeding-related Crashes', color='abbrev', size='total')
fig2.show()
```

**Interpretation:**

The scatter plot would illustrate the relationship between alcohol-related and speeding-related crashes for each state. Each state would be represented as a point, with its position determined by the number of alcohol-related crashes (on the x-axis) and speeding-related crashes (on the y-axis). The color would represent the state abbreviation, while the size of each point would be proportional to the total number of crashes. This visualization can help identify states where alcohol or speeding is a predominant factor in car crashes.

```python
# plot 3: Choropleth Map of Total Crashes by State
fig3 = px.choropleth(df, locations='abbrev', locationmode='USA-states', color='total', scope="usa", title='Total Car Crashes by State')
fig3.show()
```

**Interpretation:**

The choropleth map would provide a geographical representation of the total number of car crashes across the United States. Each state's color intensity would correspond to the number of crashes, with darker shades indicating higher crash numbers. This visualization can help identify states or regions with particularly high or low numbers of car crashes.

```python
# plot 4: Scatter Plot Matrix (SPLOM) for Numeric Columns
fig4 = px.scatter_matrix(df, dimensions=['alcohol', 'speeding', 'ins_losses', 'total'], title='Scatter Plot Matrix of Car Crash Factors')
fig4.show()
```

**Interpretation:**

The Scatter Plot Matrix (SPLOM) would display pairwise scatter plots for the selected numeric columns (alcohol, speeding, ins_losses, and total). This matrix allows for a comprehensive view of the relationships between these factors. For example, if there's a strong positive correlation between alcohol and total crashes, the corresponding scatter plot would show a clear upward trend. The diagonal plots typically showcase histograms of individual columns, providing insights into the distribution of each factor.

```python
# plot 6: Boxplot of Total Crashes Grouped by No Previous Accidents
fig6 = px.box(df, x='no_previous', y='total', title='Total Crashes Grouped by No Previous Accidents', points="all")
fig6.show()
```

**Interpretation:**

## Data Visualization on mpg dataset with `altair` library

```python
# load the dataset using seaborn
df = sns.load_dataset('mpg')
df.to_csv('mpg.csv')
```

The mpg dataset consists of several columns:

1. **Unnamed: 0:** An index or identifier.
2. **mpg:** Miles per gallon.
3. **cylinders:** Number of cylinders in the engine.
4. **displacement:** Engine displacement in cubic inches.
5. **horsepower:** Engine horsepower.
6. **weight:** Weight of the car.
7. **acceleration:** Acceleration (0-60 mph time).
8. **model_year:** Model year of the car.
9. **origin:** Country of origin.
10. **name:** Name or brand of the car.

```python
import altair as alt

# Plot 1: Bar chart of average mpg by number of cylinders
plot1 = alt.Chart(df).mark_bar().encode(
    x='cylinders:O',
    y='average(mpg):Q',
    color='cylinders:N'
).properties(
    title="Average MPG by Number of Cylinders",
    width=600,   # width in pixels
    height=400   # height in pixels
)
plot1
```

**Interpretation:**

**Average MPG by Number of Cylinders:**\
A bar chart that displays the average miles per gallon (MPG) for each number of cylinders. Cars with fewer cylinders tend to have higher MPG, indicating better fuel efficiency.

```python
# Plot 2: Scatter plot of horsepower vs. mpg
plot2 = alt.Chart(df).mark_circle().encode(
    x='horsepower:Q',
    y='mpg:Q',
    color='origin:N',
    tooltip=['name', 'horsepower', 'mpg']
).properties(
    title="Horsepower vs. MPG colored by Origin",
    width=600,   # width in pixels
    height=400  # height in pixels
)  
plot2
```

**Interpretation**:

**Horsepower vs. MPG colored by Origin:**\
A scatter plot illustrating the relationship between a car's horsepower and its MPG. Cars from different origins (USA, Europe, Asia, etc.) are colored differently. Typically, higher horsepower might be associated with lower MPG because powerful engines can consume more fuel.

```python
# Plot 3: Histogram of mpg distribution
plot3 = alt.Chart(df).mark_bar().encode(
    x=alt.X("mpg:Q", bin=True),
    y='count()'
).properties(
    title="Distribution of MPG",
    width=600,   # width in pixels
    height=400   # height in pixels
)
plot3
```

**Interpretation:**

**Distribution of MPG:**\
A histogram showing how MPG values are distributed across the dataset. This gives an idea of the most common MPG values and the range of MPG values in the dataset.



```python

# Plot 4: Line chart of average mpg over the years
plot4 = alt.Chart(df).mark_line().encode(
    x='model_year:O',
    y='average(mpg):Q'
).properties(
    title="Average MPG over the Years",
    width=600,   # width in pixels
    height=400   # height in pixels
)
plot4
```

**Interpretation:**

**Average MPG over the Years:**\
A line chart depicting how the average MPG has changed over the years. This can provide insights into improvements in fuel efficiency over time.


```python

# Plot 5: Bar chart of car counts by origin
plot5 = alt.Chart(df).mark_bar().encode(
    x='origin:N',
    y='count():Q',
    color='origin:N'
).properties(
    title="Car Counts by Origin",
    width=600,   # width in pixels
    height=400   # height in pixels
)
plot5
```

**Interpretation:**

**Car Counts by Origin:**\
A bar chart showing the number of cars from each origin (country) in the dataset. This provides a breakdown of the dataset by the country of origin. It shows highesy number of cars by USA followed by Japan and Europe

```python

# Plot 6: Scatter plot of weight vs. mpg
plot6 = alt.Chart(df).mark_circle().encode(
    x='weight:Q',
    y='mpg:Q',
    color='origin:N',
    tooltip=['name', 'weight', 'mpg']
).properties(
    title="Weight vs. MPG colored by Origin",
    width=600,   # width in pixels
    height=400   # height in pixels
)
plot6
```

**Interpretation:**

**Weight vs. MPG colored by Origin:**\
A scatter plot illustrating the relationship between a car's weight and its MPG. Heavier cars might generally have lower MPG because they require more energy to move. The points are colored based on the car's origin

```python

# Plot 7: Box plot of mpg by origin
plot7 = alt.Chart(df).mark_boxplot().encode(
    x='origin:N',
    y='mpg:Q'
).properties(
    title="MPG by Origin (Boxplot)",
    width=600,   # width in pixels
    height=400   # height in pixels
)
plot7
```

**Interpretation:**

**MPG by Origin (Boxplot):**\
A box plot showing the distribution of MPG values for cars from different origins. This helps in comparing the central tendency and spread of MPG values across origin


```python

# Plot 8: Scatter plot of acceleration vs. mpg
plot8 = alt.Chart(df).mark_circle().encode(
    x='acceleration:Q',
    y='mpg:Q',
    color='origin:N',
    tooltip=['name', 'acceleration', 'mpg']
).properties(
    title="Acceleration vs. MPG colored by Origin",
    width=600,   # width in pixels
    height=400   # height in pixels
)
plot8
```

**Interpretation:**

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.transform import factor_cmap
from bokeh.palettes import Category20c
from bokeh.layouts import gridplot

output_notebook()

# 1. Bar chart: Distribution of penguins by species
species_count = df['species'].value_counts().reset_index()
species_count.columns = ['species', 'count']

p1 = figure(x_range=species_count['species'].tolist(), title="Distribution of Penguins by Species", toolbar_location=None, tools="")
p1.vbar(x='species', top='count', width=0.7, source=species_count,
        line_color='white', fill_color=factor_cmap('species', palette=Category20c[len(species_count)], factors=species_count['species'].tolist()))

p1.xaxis.axis_label = "Species"
p1.yaxis.axis_label = "Count"
p1.xgrid.grid_line_color = None
p1.y_range.start = 0

show(p1)
```
